```python
detector_cars = cv2.CascadeClassifier('/content/drive/MyDrive/Colab Notebooks/Cascades/cars.xml')
imagem = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/car.jpg')
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
deteccoes = detector_cars.detectMultiScale(imagem_cinza, scaleFactor = 1.03, minNeighbors=5)
for (x, y, w, h) in deteccoes:
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 255, 0), 2)
cv2_imshow(imagem)
```

```python
detector_cars = cv2.CascadeClassifier('/content/drive/MyDrive/Colab Notebooks/Cascades/fullbody.xml')
imagem = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Images/people3.jpg')
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
deteccoes = detector_cars.detectMultiScale(imagem_cinza, scaleFactor = 1.05, minNeighbors = 5, minSize=(50, 50))
for (x, y, w, h) in deteccoes:
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 255, 0), 2)
cv2_imshow(imagem)
```
